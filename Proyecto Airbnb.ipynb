<h2> 1. Importar/Abrir datos del archivo CSV.</h2>

```python
import pandas as pd

Listings = pd.read_csv("Listings.csv", encoding="ISO-8859-1", low_memory=False)
```

```python
### Imprimimos las primeras primeras filas para conocer su contenido
Listings.head()
```

<h2>2. Emitir cualquier columna de fecha como formato de fecha y hora.</h2>

```python
### Imprimo la informacion de los tipos de datos que contienen
Listings.info()
```

```python
### Realizo la convecion de (host_since)  a formato datatime.

Listings["host_since"] = pd.to_datetime(Listings["host_since"])

Listings.info()
```

<h2>3. Filtre los datos en filas donde la ciudad sea París y mantenga solo las columnas "host_since", "vecindario", "ciudad", "alojamiento" y "precio" en su tabla.</h2>

```python
### Filtre los datos en filas donde la ciudad sea París y mantenga,
###solo las columnas "host_since", "vecindario", "ciudad", "alojamiento" y "precio" en su tabla.
```

```python
paris_Listings = Listings.query("city == 'Paris'")
paris_Listings.info()
```

```python
paris_Listings = (
    Listings
    .query("city == 'Paris'")
    .loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]]
)
paris_Listings.info()
```

```python
### Imprimimos los valores faltantes para hots_since
paris_Listings.isna().sum()
```

<h2>1. Cree un DataFrame llamado paris_listinings_neighbourhood. Agrupa los listados de parís por neighbourhood y calcula el precio medio de cada uno. ordenar por precio en orden ascendente.</h2>

```python
paris_listinings_neighbourhood = (
    paris_Listings
    .groupby("neighbourhood")
    .agg({"price" : "mean"})
    .sort_values("price")
)

paris_listinings_neighbourhood.tail()
```

<h2>2. Cree un DataFrame llamado paris_listings_accomodations. Filtra tus datos hasta el neighbourhood más caro de París. Agrúpalo por alojamientos y calcula el precio medio de cada uno. Ordenar por precio en orden ascendente.</h2>

<h2>1. Filtramos el Barrio con mayor valor y sacamos el promedio de precios, agrupando por la columna de alojamiento y utilizamos valores de ordenacion para ordenar en orden ascendente.</h2>

```python
paris_listinings_accommodates = (
    paris_Listings
    .query("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price" : "mean"})
    .sort_values("price")
)

paris_listinings_accommodates.head()
```

<h2>Imprimimos la cola</h2>

```python
paris_listinings_accommodates = (
    paris_Listings
    .query("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price" : "mean"})
    .sort_values("price")
)
paris_listinings_accommodates.tail()
```

<H2>Listados de Paris a lo largo del tiempo</H2>

```python
paris_listinings_over_time = (
    paris_Listings
    .set_index("host_since")
    .resample("YE")
    .agg({
        "neighbourhood": "count",
        "price": "mean"
    })
)

paris_listinings_over_time.head()
```
